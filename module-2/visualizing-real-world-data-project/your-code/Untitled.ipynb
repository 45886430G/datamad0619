```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies_clean
```

```python
companies_max10year = db.collection_companies_clean.find ({'$and': [{ 'founded_year': { '$gte': 2000 } }, { 'total_money_raised': { '$gte': 1000000.0 } }]})
```

```python
df = pd.DataFrame(companies_max10year)
```

```python
df['founded_year'] = df['founded_year'].fillna(0.0).astype(int)
```
