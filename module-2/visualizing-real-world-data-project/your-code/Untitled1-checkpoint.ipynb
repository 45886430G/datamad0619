```python
import pandas as pd
import re
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
dbc = db.companies.find()
```

```python
at_least_1_office = db.companies.find({"offices":{"$not":{"$size":0}}})
```

```python
df = pd.DataFrame(at_least_1_office)
```

```python
df.drop(['acquisition','created_at', 'crunchbase_url', 'homepage_url', 'image', 'partners', 'providerships', 'updated_at',  'permalink', 'products', 'external_links', 'acquisitions', 'alias_list', 'blog_feed_url', 'blog_url', 'category_code', 'deadpooled_day', 'deadpooled_month', 'deadpooled_url', 'deadpooled_year', 'email_address', 'founded_day', 'founded_month', 'phone_number', 'screenshots', 'tag_list', 'twitter_username', 'video_embeds', 'relationships', 'milestones', 'investments', 'funding_rounds', 'competitions'] , axis=1, inplace=True)
```

```python
df= df[['_id', 'description', 'founded_year', 'name', 'ipo', 'overview', 'total_money_raised', 'offices', 'number_of_employees']].fillna('Unknown')
```

```python
def get(d):
    d = d['offices']
    
    for i in d:
        if i['latitude'] and i['longitude']:
            return {"lat": d[0]['latitude'], "lng": d[0]['longitude']}


Geoloc = df[["offices"]].apply(get, result_type="expand", axis=1)
```

```python
DataBase = pd.concat([df, Geoloc], axis=1)
```

```python
DataBase.dropna()
```

```python
DataBase.drop(['offices'] , axis=1, inplace=True)
```

```python
def showmedamoney (m):
    floatmoney = float((re.findall('\d+.?\d+', m))[0])
    if m.endswith('M'):
        mill = floatmoney*1000000
        return mill
    elif m.endswith('B'):
        bill = floatmoney*1000000000
        return bill
    elif m.endswith('k'):
        thous = floatmoney*1000
        return thous
    else:
        return m
DataBase['total_money_raised']= DataBase['total_money_raised'].apply(showmedamoney)
```
