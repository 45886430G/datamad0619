```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
dbc = db.companies.find()
```

```python
at_least_1_office = db.companies.find({"offices":{"$not":{"$size":0}}})
```

```python
df = pd.DataFrame(at_least_1_office)
```

```python
df.drop(['acquisition','created_at', 'crunchbase_url', 'homepage_url', 'image', 'partners', 'providerships', 'updated_at',  'permalink', 'products', 'external_links', 'acquisitions', 'alias_list', 'blog_feed_url', 'blog_url', 'category_code', 'deadpooled_day', 'deadpooled_month', 'deadpooled_url', 'deadpooled_year', 'email_address', 'founded_day', 'founded_month', 'ipo', 'number_of_employees', 'phone_number', 'screenshots', 'tag_list', 'twitter_username', 'video_embeds'] , axis=1, inplace=True)
```

```python
df= df[['_id', 'competitions', 'description', 'founded_year', 'funding_rounds', 'investments', 'milestones', 'name', 'offices', 'overview', 'relationships', 'total_money_raised']].fillna('Unknown')
```

```python
def get(d):
    d = d['offices']
    
    for i in d:
        if i['latitude'] and i['longitude']:
            return {"lat": d[0]['latitude'], "lng": d[0]['longitude']}


Geoloc = df[["offices"]].apply(get, result_type="expand", axis=1)
```

```python
DataBase = pd.concat([df, Geoloc], axis=1)
```
