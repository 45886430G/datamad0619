```python
SharknoNa = SharksBBDD[['Case Number', 'Date', 'Year', 'Type','Country', 'Area', 'Location', 'Activity', 'Sex ', 'Age', 'Fatal (Y/N)']].fillna('Unknown')
SharknoNa
```

```python
#LIMPIEZA DE LA VARIABLE DATE
```

```python
def findmes (m):
    m = str(m)
    if len(re.findall('\-\w+\-',m))==0:
        return '-'
    else:
        t= re.findall('\-\w+\-',m)
        return t[0][1:4]
```

```python
SharknoNa['Date']= SharknoNa['Date'].apply(findmes)

display(SharknoNa['Date'])
```

```python
SharknoNa['Date'].replace(["Ap-",'Jut', '26-', '30-'], "-")

display(SharknoNa)
```

```python
SharknoNa=SharknoNa[SharknoNa['Date']!='-']
display(SharknoNa)
```

```python
#LIMPIEZA AGE
```

```python
def findage (m):
    
    m = str(m)
    b = re.findall ('\d+', m)
    
    if len(b) == 1:
        return b[0]
    
    else:
        return '-'
```

```python
SharknoNa['Age']= SharknoNa['Age'].apply(findage)

display(SharknoNa['Age'])
```

```python
SharknoNa=SharknoNa[SharknoNa['Age']!='-']
display(SharknoNa)
```

```python
#Limpieza de FATAL
```

```python
def findFATAL (m):
    
    m = str(m)
    b = re.findall ('N|Y', m)
    
    if len(b) == 1:
        return b[0]
    
    else:
        return '-'
```

```python
SharknoNa['Fatal (Y/N)']= SharknoNa['Fatal (Y/N)'].apply(findFATAL)

display(SharknoNa['Fatal (Y/N)'])
```

```python
print(set(SharknoNa['Fatal (Y/N)']))
```

```python
SharknoNa=SharknoNa[SharknoNa['Fatal (Y/N)']!='-']
display(SharknoNa)
```
