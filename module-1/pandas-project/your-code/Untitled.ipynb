```python
print((SharknoNa['Date']))
```

```python
def findmes (m):
    m = str(m)
    if len(re.findall('\-\w+\-',m))==0:
        return '-'
    else:
        t= re.findall('\-\w+\-',m)
        return t[0][1:4]
```

```python
SharknoNa['Date']= SharknoNa['Date'].apply(findmes)

display(SharknoNa['Date'])
```

```python
SharknoNa['Date'].value_counts()
```

```python
SharknoNa['Date'].replace(["Ap-",'Jut', '26-', '30-'], "-")

display(SharknoNa)
```

```python
def findage (a):
    a = str(a)
    b = re.findall('\d+', a)
    if len(b) == 1:
        return b[0]
    else:
        return '-'
```

```python
SharknoNa['Age']= SharknoNa['Age'].apply(findage)

display(SharknoNa['Age'])
```
