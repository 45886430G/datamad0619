#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/search?tableCount%5B%5D=0-10&tableCount%5B%5D=10-30&dataType%5B%5D=Numeric&databaseSize%5B%5D=KB&databaseSize%5B%5D=MB)

```python
from sqlalchemy import create_engine

mysql_domain = "relational.fit.cvut.cz"
mysql_user = "guest"
mysql_password = 'relational'
mysql_db = "stats"

dbdata = [mysql_user, mysql_password, mysql_domain, mysql_db]
connection_url = 'mysql+pymysql://{}:{}@{}/{}'.format(*dbdata)
engine = create_engine(connection_url)
print("Connected to DB!")
```

#### 4. Import the users table 

```python
sqlquery = """
SELECT * FROM users
"""

users = pd.read_sql_query(sqlquery, engine)

users.head()

```

#### 5. Rename Id column to userId

```python
users.rename(columns = {'Id':'UserId'}, inplace=True)
users.head()
```

#### 6. Import the posts table. 

```python
sqlquery = """
SELECT * FROM posts
"""

posts = pd.read_sql_query(sqlquery, engine)

posts.head()

```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename(columns = {'Id':'postId'}, inplace=True)
posts.head()

posts.rename(columns = {'OwnerUserId':'UserId'}, inplace=True)
posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
x = pd.DataFrame(users, columns = ['UserId', 'Reputation','Views','UpVotes','DownVotes'])
display(x.head())
```

```python
b = pd.DataFrame(posts, columns = ['postId', 'Score','UserId','ViewCount','CommentCount'])
display(b.head(10))
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_df = pd.merge(x, b, 'left')
merge_df.head()
print(merge_df.shape)
```
